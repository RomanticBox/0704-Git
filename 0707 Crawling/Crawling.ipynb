```python
# Bring Libraries
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
url = 'https://www.musinsa.com/app/codimap/lists?style_type=&tag_no=&brand=&display_cnt=60&list_kind=big&sort=comment_cnt&page=1'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}

res = requests.get(url, headers=headers)
res.raise_for_status()
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
filename = 'crawling.csv'
f = open(filename, 'w', encoding='cp949', newline='')
writer = csv.writer(f)
```

```python
FashionBox = soup.find('ul', attrs = {'class' : 'style-list'})
fashions = FashionBox.find_all('li')
codimap_category = []
codimap_title = []
codimap_date = []
views = []
comment_numbers = []
```

```python
#카테고리, 이름, 날짜, 조회수, 댓글 수
for fashion in fashions:
    category = fashion.find('span', attrs = {'class' : 'style-list-information__text'}).get_text()
    title = fashion.find('strong', attrs = {'class' : 'style-list-information__title'}).get_text()
    post_infos = fashion.find_all('span', attrs = {'class' : 'post-information__text'})
    date = str(post_infos[0])
    date = date[37:45]
    view = str(post_infos[1])
    view = view[40:-7]
    reply = str(post_infos[2])
    reply = reply[40:-7]
    while reply[0] not in ['1', '2', '3', '4', '5', '6', '7', '8', '9']:
        reply = reply.strip(reply[0])
    print(len(reply))
    print(reply)
    codimap_category.append(category)
    codimap_title.append(title)
    codimap_date.append(date)
    views.append(view)
    comment_numbers.append(reply)
```

```python
data_title = ['codimap_category', 'codimap_title', 'codimap_date', 'views', 'comment_numbers']
writer.writerow(data_title)
for i in range(len(views)):
    new = [codimap_category[i], codimap_title[i], codimap_date[i], views[i], comment_numbers[i]]
    writer.writerow(new)
f.close()
```
